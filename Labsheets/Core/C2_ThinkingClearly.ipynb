At this stage in the coursework you will begin to integrate your code into a final solution.  This will mean that your robot will need to reliably operate several behaviours and to switch between them appropriately.  This has the unfortunate side effect that any bugs in your code will be much harder to track down and to solve.  

It is also true that, as you begin to test your final solution, it will simply take longer to evaluate.  Consider:
- if your Romi takes 3 minutes to complete one full test of the line challenge...
- ...and 5 minutes to edit and reprogram each time...
- ...and your bug takes 50 attempts to track down...
- ...you will spend approximately **6 hours 40 minutes** watching your robot do the wrong thing.

 $$:($$

At this stage it may be very tempting to keep attacking the problem with lots of energy.  However, the best advice is to **work smart, not hard**.  Sometimes it is actually better to:
- take a break.
- think about the problem clearly.
- and talk to someone about it.  

Look back over the labsheets Check Point 0, Check Point 1, and Labsheet 0: Troubleshooting, and double check that you are following the advice.  It will make your life easier.

If you have an idea of where the bug is in your system, it is much better (quicker, in the long run) to: 
- Construct a small test scenario
- Isolate the code as best as possible (perhaps even write the problem code into a new small program).
- Refactor your code.

Hopefully you have had some experience with constructing small experiments when developing your PID controllers.  

In terms of the **intended learning outcomes** of this unit, it is at this stage of Assessment 1 that you will need to exercise your **post-graduate** level skills of critical reasoning, analysis, evaluation and synthesis.  You are highly encouraged to discuss the bugs you find with your peers to develop your ability to communicate clearly and critically.  Describing your problem can go a long way to solving it.  It would be useful to identify your errors and bugs as **syntax, execution or logic**.   Of course, please bring your observations to the teaching staff for discussion.


# Looking Ahead

## PID

Labsheet 7 covers PID control.  When PID goes wrong your robot will behave very irratically.  It can be difficult to work out what is causing the problem.  Therefore, follow the guidance in the labsheet to work on one element of the PID controller at a time.   Tuning PID will be a good test of both your software debugging skills, and your ability to diagnostically observe a robotic system.

It can also be useful to "think outside the box".  For example, start by placing your Romi on top of a cup so that it is not actually moving around.  This will reduce the time you spend chasing after your robot. 


## Kinematics

Labsheet 8 covers kinematics for your Romi, which will allow you to track the position of your Romi relative to where it powered-up.  This functionality is useful for:
- driving reliably in a straight line.
- driving a fixed distance.
- knowing how far you have travelled in total (e.g., being sure of the line end).
- turning to face a direction reliably.
- driving back to the start location.

However, you will discover that the performance of your kinematics will depend upon:
- The motion control of your wheels.
- Your strategies for addressing the challenges along the line.
- The calibration of the physical parameters of your Romi.

In each of these cases, it is best to construct small test scenarios to evaluate the performance and to calibrate parameters.  When you are developing your kinematics, it isn't necessary for your Romi to also exhibit all of the other behaviours for the line following challenge.


## Finite State Machine

At this stage you may have a lot of code which fits together in a way that has grown out of the labsheets.  It is strongly recommended you take the time to implement a Finite State Machine (or similar), which will require you to refactor your code.  Utilising a finite state machine will save you time in the long run because:
- you can quickly enable or disable behaviours.
- you can quickly change criteria and selection of behaviours.
- you can quickly change the order of sequence of behaviours.
- you can quickly create new, slightly modified behaviours, and integarate them.
- your code will be easier to read, and therefore, it will be easier to spot the mistakes.
- it will be easier to keep track of key variables, and when they might need to be reinitialised.
